# Part 2: SQL Queries (10 Queries)

Write SQL queries to answer these business questions. Run each query using Python's
pd.read_sql(query, conn) and show the output.


```python
# Query 1: [Show employee names, salaries, and cities sorted by salary (highest first)]

query_1 = """
SELECT
    name,
    salary,
    city
FROM employees
ORDER BY salary DESC;
"""

result_1 = pd.read_sql(query_1, conn)
print(result_1)
```

```python
# Query 2: [Find all employees in Mumbai or Bangalore with salary > 50000]

query_2 = """
SELECT
    name,
    salary,
    city
FROM employees
WHERE (city = 'Mumbai' OR city = 'Bangalore')
  AND salary > 50000
ORDER BY salary DESC;
"""

result_2 = pd.read_sql_query(query_2, conn)
print(result_2)
```

```python
# Query 3: [Find all employees whose names start with 'A' or 'S']

query_3 = """
SELECT
    name,
    salary,
    city
FROM employees
WHERE name LIKE 'A%'
   OR name LIKE 'S%'
ORDER BY name;
"""

result_3 = pd.read_sql_query(query_3, conn)
print(result_3)
```

```python
# Query 4:[Show each employee's name along with their department name. (Join employees + departments)]

query_4 = """
SELECT
    employees.name,
    departments.dept_name
FROM employees
INNER JOIN departments
ON employees.dept_id = departments.dept_id;
"""

result_4 = pd.read_sql_query(query_4, conn)
print(result_4)
```

```python
# Query 5:[Show ALL departments and the number of employees in each department. Include departments with zero employees.]

query_5 = """
SELECT
    departments.dept_name,
    COUNT(employees.emp_id) AS employee_count
FROM departments
LEFT JOIN employees
ON departments.dept_id = employees.dept_id
GROUP BY departments.dept_id, departments.dept_name;
"""

result_5 = pd.read_sql_query(query_5, conn)
print(result_5)
```

```python
# Query 6:[Show each employee's name, their department name, and their assigned project name. (Join all 3tables)]

query_6 = """
SELECT
    employees.name,
    departments.dept_name,
    projects.project_name
FROM employees
INNER JOIN departments
ON employees.dept_id = departments.dept_id
INNER JOIN projects
ON employees.emp_id = projects.emp_id;
"""

result_6 = pd.read_sql_query(query_6, conn)
print(result_6)
```

```python
# Query 7:[Find the average salary, total salary expense, and employee count for each department.]

query_7 = """
SELECT
    departments.dept_name,
    AVG(employees.salary) AS average_salary,
    SUM(employees.salary) AS total_salary,
    COUNT(employees.emp_id) AS employee_count
FROM departments
INNER JOIN employees
ON departments.dept_id = employees.dept_id
GROUP BY departments.dept_id, departments.dept_name;
"""

result_7 = pd.read_sql_query(query_7, conn)
print(result_7)
```

```python
# Query 8:[Find departments where the average salary is greater than ₹60,000.]

query_8 = """
SELECT
    departments.dept_name,
    AVG(employees.salary) AS average_salary
FROM departments
INNER JOIN employees
ON departments.dept_id = employees.dept_id
GROUP BY departments.dept_id, departments.dept_name
HAVING AVG(employees.salary) > 60000;
"""

result_8 = pd.read_sql_query(query_8, conn)
print(result_8)
```

```python
# Query 9:[Find all employees who earn more than the company-wide average salary]

query_9 = """
SELECT
    name,
    salary
FROM employees
WHERE salary > (
    SELECT AVG(salary)
    FROM employees
);
"""

result_9 = pd.read_sql_query(query_9, conn)
print(result_9)
```

```python
# Query 10:[Find the employee with the 2nd highest salary in the company. Show their name, department, andsalary.]

query_10 = """
SELECT
    employees.name,
    departments.dept_name,
    employees.salary
FROM employees
INNER JOIN departments
ON employees.dept_id = departments.dept_id
WHERE employees.salary = (
    SELECT MAX(salary)
    FROM employees
    WHERE salary < (
        SELECT MAX(salary)
        FROM employees
    )
);
"""

result_10 = pd.read_sql_query(query_10, conn)
print(result_10)
```

# Part 3: Python Analysis (Pandas + NumPy)

Load the data from your database into Pandas DataFrames and perform the following analysis tasks
using Python.


```python
# A. Load Data into DataFrames
# Step 1: Load each table

df_emp = pd.read_sql("SELECT * FROM employees", conn)
df_dept = pd.read_sql("SELECT * FROM departments", conn)
df_proj = pd.read_sql("SELECT * FROM projects", conn)

# Step 2: Verify by printing first few rows
print(df_emp.head())
print(df_dept.head())
print(df_proj.head())
```

```python
# B. Data Overview
print(df_emp.shape)             # (rows, columns)
print(df_emp.info())            # column types & non-null counts
print(df_emp.describe())        # statistics
print(df_emp.head())            # first 5 rows
print(df_emp.isnull().sum())    # missing values
```

```python
# C. NumPy Operations
# Step 1: Convert salary column to NumPy array
salaries = np.array(df_emp['salary'])

# Step 2: Compute statistics
print("Mean:", np.mean(salaries))
print("Median:", np.median(salaries))
print("Std Dev:", np.std(salaries))
print("Min:", np.min(salaries))
print("Max:", np.max(salaries))

# Step 3: Vectorized annual salary
annual = salaries * 12
print("Annual salaries:")
print(annual)
```

```python
# D. Pandas GroupBy
print("Average Salary by City")
print(df_emp.groupby('city')['salary'].mean())
merged = pd.merge(df_emp, df_dept, on='dept_id')

print("Total Salary Expense by Department")
print(merged.groupby('dept_name')['salary'].sum())
print("City with Highest Average Salary")
print(df_emp.groupby('city')['salary'].mean().idxmax())
```

```python
# E. Pandas Merge (= SQL JOIN)
# Merge on dept_id (same as INNER JOIN)

merged = pd.merge(df_emp, df_dept, on='dept_id')

print(merged[['name', 'dept_name']])

# Compare: This matches SQL Query 4
```
